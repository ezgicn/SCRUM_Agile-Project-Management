```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the burndown chart
iterations = ["I1", "I2", "I3", "I4", "I5", "I6", "I7", "I8"]
total_effort = 160
team_velocity = 20  # Effort points per iteration
work_done_per_iteration = [20, 20, 20, 20, 20, 20, 20, 20]
```

```python
# Calculate work remaining
work_remaining = [total_effort - sum(work_done_per_iteration[:i]) for i in range(len(iterations)+1)]

# Calculate trendline for ideal burndown
ideal_burndown = np.linspace(total_effort, 0, len(iterations) + 1)

# Create the burndown chart
plt.figure(figsize=(10, 6))
```

```python
# Plot work remaining
plt.plot(range(len(work_remaining)), work_remaining, label="Actual Work Remaining", marker="o")

# Plot ideal burndown
plt.plot(range(len(ideal_burndown)), ideal_burndown, label="Ideal Work Remaining", linestyle="--", color="green")

# Highlight current iteration
current_iteration = 4
plt.scatter(current_iteration, work_remaining[current_iteration], color="red", label="Current Iteration", zorder=5)
plt.annotate("You are here", (current_iteration, work_remaining[current_iteration]),
             textcoords="offset points", xytext=(-50, -20), ha="center", color="red", fontsize=10)

# Add labels and title
plt.xticks(range(len(work_remaining)), ["Start"] + iterations)
plt.yticks(range(0, total_effort + 20, 20))
plt.xlabel("Iterations")
plt.ylabel("Work Remaining (Effort Points)")
plt.title("Sprint Burndown Chart")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the chart
plt.show()
```
